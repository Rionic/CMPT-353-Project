```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn
seaborn.set()

# Getting data into a form we want
df = pd.read_csv('business_reviews_users_merged.csv')
df = df.drop(columns=['Unnamed: 0', 'business_id', 'business_name', 'is_open', 'business_stars', 'business_review_count', 'elite', 'review_id'])

# Filter out unknown id name
df = df[df['user_id'] != '#NAME?']

# Sort by day of review for each user
df = df.groupby(['user_id', 'date']).first().reset_index()
df['review_order'] = df.groupby(['user_id']).cumcount()+1

# Average rating of each user by user review order
avg = df.drop(columns=['user_id', 'date', 'average_stars', 'username', 'user_review_count'])
avg = avg.groupby(['review_order']).mean().reset_index()
# avg = avg[avg['review_order'] < 300]

fit = stats.linregress(avg['review_order'], avg['review_stars'])
avg['prediction'] = avg['review_order']*fit.slope + fit.intercept

plt.plot(avg['review_order'], avg['review_stars'], 'b.', alpha=1.0)
plt.plot(avg['review_order'], avg['prediction'], 'r-', linewidth=3)
plt.show()
```
